```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,r2_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:\\Users\\user\\saurabh sir pandas\\Final Projects\\Disease_symptom_and_patient_profile_dataset.csv")
```

```python
df=df.sample(2000,replace=True)
df
```

```python
df.isnull().sum()
```

```python
lr=LabelEncoder()
oe=OrdinalEncoder()
```

```python
df["Fever"]=lr.fit_transform(df["Fever"])
df["Cough"]=lr.fit_transform(df["Cough"])
df["Fatigue"]=lr.fit_transform(df["Fatigue"])
df["Difficulty Breathing"]=lr.fit_transform(df["Difficulty Breathing"])
df["Gender"]=lr.fit_transform(df["Gender"])
df["Blood Pressure"]=lr.fit_transform(df["Blood Pressure"])
df["Cholesterol Level"]=lr.fit_transform(df["Cholesterol Level"])
df["Outcome Variable"]=lr.fit_transform(df["Outcome Variable"])
# df["Disease"]=lr.fit_transform(df["Disease"])
```

```python
df 
```

```python
from  sklearn.preprocessing import StandardScaler
```

```python
sc= StandardScaler() 
```

```python
x=df.drop(columns="Disease")
y=df["Disease"]
```

```python
x
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
rf=RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```
